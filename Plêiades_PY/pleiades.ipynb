```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data_pleiades.csv", sep = ',' , engine= 'python')
```

```python
data.head()
```

```python
len(data)
```

```python
plt.style.use('dark_background')
plt.rc('font', size = 15)
```

```python
plt.figure(figsize= (10,8))
plt.scatter(data['ra'], data['dec'], s=5, alpha=0.01, color= 'blue')

plt.xlabel('RA')
plt.ylabel('DEC')
plt.show()
```

```python
data_r = data.loc[data['g_rp']<0.6]
```

```python
len(data_r)
```

```python
plt.figure(figsize = (10,8))
plt.scatter(data_r['pmra'], data_r['pmdec'], s= 5 , alpha= 0.1, color='blue')
plt.xlabel('pmra')
plt.ylabel('pmdec')
plt.show()
```

```python
comoving = data.loc[data['pmra'] <= 23 ]
comoving = comoving.loc[comoving['pmra'] >= 17 ]
comoving = comoving.loc[comoving['pmdec'] <= -42 ]
comoving = comoving.loc[comoving['pmdec'] >= -50 ]
```

```python
plt.figure(figsize=(10,8))
plt.hist(comoving['parallax'], bins = 80, color= 'blue', alpha = 0.8)
plt.xlabel('parallax')
plt.ylabel('numbers')
plt.show()
```

```python
pleiades = comoving.loc[comoving['parallax'] <= 9.0]
pleiades = pleiades.loc[pleiades['parallax'] >= 6.0]
```

```python
p = np.mean(pleiades['parallax'])
p
```
